```python
import pandas as pd
import seaborn as sns
import yaml
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("../Data/clean/df_final_web_data.csv")
# convert time column
df1['date_time'] = pd.to_datetime(df1['date_time'])
# sort events in correct order
df1 = df1.sort_values(['visit_id','date_time'])
df2 = pd.read_csv("../Data/clean/df_final_experiment_clients_clean.csv")
```

```python
df2 = pd.read_csv("../Data/clean/df_final_experiment_clients_clean.csv")
df = pd.merge(df1, df2, on="client_id", how="inner")
df
```

```python
def calculate_completion_rate(df):
    """
    Calculate completion rate per visitor_id, grouped by Variation.
    Completion = visitor reached the 'confirm' process_step.
    
    Args:
        df: DataFrame with columns [client_id, visitor_id, visit_id, process_step, date_time, Variation]
    
    Returns:
        DataFrame with completion rates per Variation
    """
    # Total unique visitors per Variation
    total = df.groupby('Variation')['visitor_id'].nunique().reset_index()
    total.columns = ['Variation', 'total_visitors']
    
    # Unique visitors who reached 'confirm'
    confirmed = (
        df[df['process_step'] == 'confirm']
        .groupby('Variation')['visitor_id']
        .nunique()
        .reset_index()
    )
    confirmed.columns = ['Variation', 'completed_visitors']
    
    # Merge and calculate rate
    result = total.merge(confirmed, on='Variation', how='left')
    result['completed_visitors'] = result['completed_visitors'].fillna(0).astype(int)
    result['completion_rate_%'] = (result['completed_visitors'] / result['total_visitors'] * 100).round(2)
    
    return result


print(calculate_completion_rate(df))
```

```python
from scipy import stats

def calculate_completion_rate_with_pvalue(df):
    """
    Calculate completion rate per visitor_id, grouped by Variation,
    and run a chi-square significance test between variations.
    
    Args:
        df: DataFrame with columns [client_id, visitor_id, visit_id, process_step, date_time, Variation]
    
    Returns:
        - summary: DataFrame with completion rates per Variation
        - chi2, p_value, significant: test results
    """
    # Total unique visitors per Variation
    total = df.groupby('Variation')['visitor_id'].nunique().reset_index()
    total.columns = ['Variation', 'total_visitors']

    # Unique visitors who reached 'confirm'
    confirmed = (
        df[df['process_step'] == 'confirm']
        .groupby('Variation')['visitor_id']
        .nunique()
        .reset_index()
    )
    confirmed.columns = ['Variation', 'completed_visitors']

    # Merge and calculate rate
    result = total.merge(confirmed, on='Variation', how='left')
    result['completed_visitors'] = result['completed_visitors'].fillna(0).astype(int)
    result['not_completed'] = result['total_visitors'] - result['completed_visitors']
    result['completion_rate_%'] = (result['completed_visitors'] / result['total_visitors'] * 100).round(2)

    # --- Chi-square test ---
    # Contingency table: [[completed, not_completed], [...]]
    contingency_table = result[['completed_visitors', 'not_completed']].values

    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

    # Print results
    print("=" * 45)
    print(result[['Variation', 'total_visitors', 'completed_visitors', 'completion_rate_%']].to_string(index=False))
    print("=" * 45)
    print(f"Chi-square statistic : {chi2:.4f}")
    print(f"P-value              : {p_value:.4f}")
    print(f"Degrees of freedom   : {dof}")
    print("-" * 45)
    if p_value < 0.05:
        print("✅ Result is SIGNIFICANT (p < 0.05)")
        print("   The difference in completion rates is unlikely due to chance.")
    else:
        print("❌ Result is NOT significant (p >= 0.05)")
        print("   No statistically significant difference found.")
    print("=" * 45)

    return result, chi2, p_value


result, chi2, p_value = calculate_completion_rate_with_pvalue(df)
```

```python
def calculate_relative_uplift(df, control_label='control', test_label='test'):
    """
    Calculate relative uplift in completion rate between test and control.
    
    Relative Uplift = ((test_rate - control_rate) / control_rate) * 100

    Args:
        df:              DataFrame with columns [visitor_id, process_step, Variation]
        control_label:   Name of the control group (default: 'control')
        test_label:      Name of the test group (default: 'test')

    Returns:
        dict with rates, absolute difference, and relative uplift
    """

    # Total unique visitors per Variation
    total = df.groupby('Variation')['visitor_id'].nunique().reset_index()
    total.columns = ['Variation', 'total_visitors']

    # Unique visitors who reached 'confirm'
    confirmed = (
        df[df['process_step'] == 'confirm']
        .groupby('Variation')['visitor_id']
        .nunique()
        .reset_index()
    )
    confirmed.columns = ['Variation', 'completed_visitors']

    # Merge and calculate rates
    result = total.merge(confirmed, on='Variation', how='left')
    result['completed_visitors'] = result['completed_visitors'].fillna(0).astype(int)
    result['completion_rate']    = result['completed_visitors'] / result['total_visitors']
    result['completion_rate_%']  = (result['completion_rate'] * 100).round(2)

    # Extract control & test rates
    r_ctrl = result.loc[result['Variation'] == control_label, 'completion_rate'].values[0]
    r_test = result.loc[result['Variation'] == test_label,    'completion_rate'].values[0]

    # Calculate uplift metrics
    absolute_diff   = (r_test - r_ctrl) * 100
    relative_uplift = ((r_test - r_ctrl) / r_ctrl) * 100

    # Print report
    print("=" * 45)
    print("        RELATIVE UPLIFT ANALYSIS")
    print("=" * 45)
    print(f"Control completion rate : {r_ctrl*100:.2f}%")
    print(f"Test completion rate    : {r_test*100:.2f}%")
    print("-" * 45)
    print(f"Absolute difference     : {absolute_diff:+.2f} pp")
    print(f"Relative uplift         : {relative_uplift:+.2f}%")
    print("-" * 45)

    if relative_uplift > 0:
        print(f"✅ Test performs {relative_uplift:.2f}% BETTER than control")
    elif relative_uplift < 0:
        print(f"❌ Test performs {abs(relative_uplift):.2f}% WORSE than control")
    else:
        print("➡️  No difference between test and control")
    print("=" * 45)

    return {
        'control_rate_%'  : round(r_ctrl * 100, 2),
        'test_rate_%'     : round(r_test * 100, 2),
        'absolute_diff_pp': round(absolute_diff, 2),
        'relative_uplift_%': round(relative_uplift, 2)
    }


uplift = calculate_relative_uplift(df)
```

**Conclusion**
**Completion Rate Analysis**
The analysis provides strong statistical evidence to reject the null hypothesis that there is no difference in completion rates between the control and test variations.
The test group achieved a completion rate of 65.19% compared to 59.24% in the control group, representing an absolute improvement of +5.95 percentage points and a relative uplift of +10.05%. This means that users exposed to the test variation were completing the process at a meaningfully higher rate than those in the control group.
The chi-square test confirms this difference is statistically significant. With a p-value effectively equal to zero, the probability of observing a difference this large by random chance alone is negligible. The size of data sample shows that the results are reliable and not a result of insufficient data.
In practical terms, a 10.05% relative uplift in completion rate is a substantial business gain. If the test variation were rolled out to all users, roughly 1 in 10 additional visitors who previously would not have completed the process now would.
Recommendation: Based on both statistical significance and practical effect size, the test variation should be adopted as the new default experience.
